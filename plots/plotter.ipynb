```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
ours = torch.load('secondOrder.pth')
them = torch.load('newton.pth')

ours = ours.cpu().numpy().reshape(-1, )
them = them.cpu().numpy().reshape(-1, )
```

```python
len(ours[ours > 0]), len(them[them > 0])
```

```python
bins = np.linspace(-0.01, 3, 100)
plt.hist(ours[ours > 0], bins=bins, alpha=0.5, label='Ours');
plt.hist(them[them > 0], bins=bins, alpha=0.5, label='Newton');
plt.legend();
plt.xlabel('Certified Radius');
plt.show()
```

```python
index = np.argsort(ours)
ourSort = np.array(ours)[index].reshape(-1, 1)
themSort = np.array(them)[index].reshape(-1, 1)

ourSort = ourSort[ourSort > 0]
themSort = themSort[themSort > 0]

plt.plot(ourSort, label='ours')
plt.plot(themSort, label='theirs')
plt.legend();
plt.xlabel('Sample');
plt.ylabel('Certified Radius');
plt.show()
```

```python
bins = np.linspace(-0.05, 0.8, 51)
plt.hist(them[them > 0] - ours[ours > 0], bins=bins, alpha=1);
plt.xlabel('Difference in Certified Radius');
plt.show()
```
